```python
# export
df = pd.DataFrame([Image.open(bss[i]).shape for i in range(len(bss))], columns=['height', 'width'])
ax = df.plot()
ax.set_title("board height and width")
df.describe()
```

```python
# export
def plot_samples(data):
    fig = plt.figure(figsize=(10, 10))
    gs = fig.add_gridspec(4, 4)
    for i in range(16):
        j = i // 4
        k = i % 4
        ax = fig.add_subplot(gs[j, k])
        ax.set_xticks([])
        ax.set_xticklabels([])
        ax.set_yticklabels([])
        ax.grid(None)
        idx = randint(0, len(data))
        ax.matshow(Image.open(data[idx]))
```

```python
plot_samples(bss)
```

```python
plot_samples(wbs)
```
