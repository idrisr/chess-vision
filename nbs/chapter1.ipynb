```python
from fastai.data.external import URLs, untar_data
from fastai.vision.data import ImageDataLoaders
from fastai.vision.learner import cnn_learner
from fastai.vision.augment import Resize
from fastai.vision.models import resnet34
from fastai.data.transforms import get_image_files
from pathlib import Path
from IPython import display
from random import randint
from fastai.metrics import error_rate
import matplotlib.pyplot as plt
from PIL import Image
from io import BytesIO
import pandas as pd
```

```python
data = Path('/home/idris/.fastai/data/oxford-iiit-pet')
ims = Path(data/'images').ls()
anno = Path(data)
```

```python
fig, ax = plt.subplots(nrows=3, ncols=3, figsize=(8, 8))
for row in range(3):
    for col in range(3):
        ax[row][col].imshow(Image.open(ims[randint(0, 7393)]))
```

```python
dls = ImageDataLoaders.from_name_func(data, 
                                      fnames=get_image_files(data/'images'), 
                                      seed=42,
                                      valid_pct=0.2,
                                      label_func=lambda y: y[0].isupper(), 
                                      item_tfms=Resize(224))
```

```python
learner = cnn_learner(dls, arch=resnet34, metrics=error_rate)
```

```python
learner.fine_tune(1)
```
