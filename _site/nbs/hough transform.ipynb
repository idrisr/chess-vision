```python
import cv2 as cv
import numpy as np
from pathlib import Path
from PIL import Image
```

```python
inFile = Path('/home/idris/2050-chess-vision/training-data-app/data/_board/maya.png')
# inFile = Path("lichess.png")
outFile = Path("houghlines.jpg")
```

```python
img = cv.imread(inFile.as_posix())
```

```python
Image.open(inFile);
```

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
Image.fromarray(gray);
```

```python
edges = cv.Canny(gray,50,150,apertureSize = 3)
```

```python
lines = cv.HoughLines(edges,1,np.pi/180,200)
len(edges)
```

```python
for line in lines:
    rho,theta = line[0]
    a = np.cos(theta)
    b = np.sin(theta)
    x0 = a*rho
    y0 = b*rho
    x1 = int(x0 + 10000*(-b))
    y1 = int(y0 + 10000*(a))
    x2 = int(x0 - 10000*(-b))
    y2 = int(y0 - 10000*(a))
    cv.line(img,(x1,y1),(x2,y2),(0,0,255),2)
cv.imwrite(outFile.as_posix(),img)
```

```python
Image.open(outFile)
```
